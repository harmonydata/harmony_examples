```python
from harmony.matching.generate_crosswalk_table import generate_crosswalk_table
```

```python
!wget https://github.com/harmonydata/harmony_examples/raw/refs/heads/main/variables_for_harmony_example.csv
```

```python
import pandas as pd
input_file_name = "variables_for_harmony_example.csv"
df = pd.read_csv(input_file_name)
```

```python
df
```

## Convert the data frame to Instruments (corresponding to sweeps)

```python
from harmony import create_instrument_from_list, match_instruments
```

```python
instruments = []
for (survey_name, year), survey_content_df in df.groupby(["survey_name", "year"]):
  this_survey = create_instrument_from_list(list(survey_content_df.item_text), list(survey_content_df.item_name), instrument_name=f"{survey_name} {year}", file_name=input_file_name)
  instruments.append(this_survey)
```

```python
print (f"We found {len(instruments)} surveys in a data frame of {len(df)} entries")
```

```python
questions, similarity, query_similarity, new_vectors_dict = match_instruments(instruments)
```

## Set the threshold for the crosswalk table

```python
threshold = 0.6
```

```python
df_crosswalk_table = generate_crosswalk_table(questions, similarity, threshold)
```
